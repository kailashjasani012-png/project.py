```python
plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(), cbar=False, cmap="viridis" )
plt.title("Missing Values Heatmap")
plt.show()
```

```python
numeric_columns = df.select_dtypes(include=['number']).columns

for col in numeric_columns:
    df[col] = df[col].fillna(df[col].mean())

categorical_columns = df.select_dtypes(include=['object']).columns

for col in categorical_columns:
    df[col] = df[col].fillna(df[col].mode()[0])

df.isnull().sum()
```

```python
df.head(10)
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.columns = df.columns.str.strip()
df.columns = df.columns.str.replace(" ", "_")
df.columns = df.columns.str.lower()

df.columns
```

```python
df.shape
```

```python
print(df.columns)
```

```python
top10 = df.sort_values(by="happiness_score", ascending=False).head(10)

plt.figure(figsize=(10,6))

sns.barplot(
    data=top10,
    x="happiness_score",
    y="country"
)

plt.title("Top 10 Happiest Countries")
plt.xlabel("Happiness Score")
plt.ylabel("Country")

plt.show()
```

```python
bottom10 = df.sort_values(by="happiness_score", ascending=True).head(10)

plt.figure(figsize=(10,6))

sns.barplot(
    data=bottom10,
    x="happiness_score",
    y="country"
)

plt.title("Bottom 10 Happiest Countries")
plt.xlabel("Happiness Score")
plt.ylabel("Country")

plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.scatterplot(
    data=df,
    x="gdp_per_capita",
    y="happiness_score",
    hue="region"
)

plt.title("GDP per Capita vs Happiness Score")
plt.xlabel("GDP per Capita")
plt.ylabel("Happiness Score")

plt.legend(bbox_to_anchor=(1.05,1), loc="upper left")
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.regplot(
    data=df,
    x="social_support",
    y="happiness_score",
    scatter_kws={"alpha":0.6}
)

plt.title("Relationship Between Social Support and Happiness Score")
plt.xlabel("Social Support")
plt.ylabel("Happiness Score")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.regplot(
    data=df,
    x="healthy_life_expectancy",
    y="happiness_score",
    scatter_kws={"alpha":0.6},
    line_kws={"color":"red"}
)

plt.title("Relationship Between Healthy Life Expectancy and Happiness Score")
plt.xlabel("Healthy Life Expectancy")
plt.ylabel("Happiness Score")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.regplot(
    data=df,
    x="freedom",
    y="happiness_score",
    scatter_kws={"alpha":0.6},
    line_kws={"color":"red"}
)

plt.title("Relationship Between Freedom and Happiness Score")
plt.xlabel("Freedom")
plt.ylabel("Happiness Score")

plt.show()
```

```python
region_avg = df.groupby("region")["happiness_score"].mean().sort_values(ascending=False)

print(region_avg)
```

```python
plt.figure(figsize=(12,6))

sns.barplot(
    x=region_avg.index,
    y=region_avg.values
)

plt.title("Average Happiness Score by Region")
plt.xlabel("Region")
plt.ylabel("Average Happiness Score")

plt.xticks(rotation=45)

plt.show()
```

```python
year_avg = df.groupby("year")["happiness_score"].mean()

print(year_avg)
```

```python
plt.figure(figsize=(10,6))

plt.plot(
    year_avg.index,
    year_avg.values,
    marker="o",
    linewidth=2
)

plt.title("Average Happiness Score by Year")
plt.xlabel("Year")
plt.ylabel("Average Happiness Score")

plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.boxplot(
    data=df,
    x="region",
    y="happiness_score"
)

plt.title("Region-wise Happiness Score Distribution")
plt.xlabel("Region")
plt.ylabel("Happiness Score")

plt.xticks(rotation=45)

plt.show()
```

```python
selected_columns = [
    "happiness_score",
    "gdp_per_capita",
    "social_support",
    "healthy_life_expectancy",
    "freedom",
    "region"
]

sns.pairplot(
    df[selected_columns],
    hue="region"
)

plt.show()
```
