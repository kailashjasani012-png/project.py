```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
```

```python
df = pd.read_csv("stock_market_data.csv")

print("Dataset Loaded Successfully")
```

```python
print("First 5 Rows")
print(df.head())
```

```python
print("\nDataset Shape")
print(df.shape)
```

```python
print("\nColumn Names")
print(df.columns)
```

```python
print("\nDataset Information")
print(df.info())
```

```python
print("\nStatistical Summary")
print(df.describe())
```

```python
print("Missing Values in Each Column")
print(df.isnull().sum())
```

```python
df["Open"] = df["Open"].fillna(df["Open"].mean())

df["Close"] = df["Close"].fillna(df["Close"].mean())

df["Volume"] = df["Volume"].fillna(df["Volume"].median())
```

```python
print(df.isnull().sum())
```

```python
df["Moving_Average_20"] = df["Moving_Average_20"].fillna(df["Moving_Average_20"].mean())

df["Moving_Average_50"] = df["Moving_Average_50"].fillna(df["Moving_Average_50"].mean())

df["Daily_Return"] = df["Daily_Return"].fillna(0)
```

```python
print(df.isnull().sum())
```

```python
print("===== Descriptive Statistics =====")

print(df.describe())
```

```python
numeric_columns = ["Open", "High", "Low", "Close", "Adj Close", "Volume"]

print(df[numeric_columns].describe())
```

```python
print("Mean Values")
print(df[numeric_columns].mean())

print("\nMedian Values")
print(df[numeric_columns].median())

print("\nStandard Deviation")
print(df[numeric_columns].std())
```

```python
df["Date"] = pd.to_datetime(df["Date"])


df = df.sort_values("Date")


plt.figure(figsize=(12,6))

plt.plot(df["Date"], df["Close"], linewidth=2)

plt.title("Closing Price Trend")
plt.xlabel("Date")
plt.ylabel("Closing Price")

plt.xticks(rotation=45)

plt.grid(True)

plt.show()
```

```python

df["Date"] = pd.to_datetime(df["Date"])


df["Year_Month"] = df["Date"].dt.to_period("M")


monthly_volume = df.groupby("Year_Month")["Volume"].mean()

plt.figure(figsize=(12,6))

plt.plot(monthly_volume.index.astype(str), monthly_volume.values, marker="o")

plt.title("Monthly Average Trading Volume")
plt.xlabel("Year-Month")
plt.ylabel("Average Volume")

plt.xticks(rotation=45)

plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(12,6))

plt.plot(df["Date"], df["Close"], label="Closing Price")

plt.plot(df["Date"], df["Moving_Average_20"], label="20-Day Moving Average")

plt.title("Closing Price vs 20-Day Moving Average")
plt.xlabel("Date")
plt.ylabel("Price")

plt.legend()

plt.xticks(rotation=45)

plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(12,6))

plt.plot(df["Date"], df["Close"], label="Closing Price")

plt.plot(df["Date"], df["Moving_Average_20"], label="20-Day MA")

plt.plot(df["Date"], df["Moving_Average_50"], label="50-Day MA")

plt.title("Closing Price with 20-Day and 50-Day Moving Averages")
plt.xlabel("Date")
plt.ylabel("Price")

plt.legend()

plt.xticks(rotation=45)

plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.histplot(
    data=df,
    x="Daily_Return",
    bins=30,
    kde=True
)

plt.title("Distribution of Daily Returns")
plt.xlabel("Daily Return")
plt.ylabel("Frequency")

plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    df[["Open","High","Low","Close","Volume","Daily_Return"]].corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Between Important Features")

plt.show()
```

```python
highest_close = df.loc[df["Close"].idxmax()]

print(highest_close)
```

```python
highest_close = df.loc[df["Close"].idxmax()]

print("===== Highest Closing Price =====")
print(highest_close)

plt.figure(figsize=(8,5))

plt.bar(highest_close["Company"], highest_close["Close"])

plt.title("Highest Closing Price")
plt.xlabel("Company")
plt.ylabel("Closing Price")

plt.show()
```

```python
lowest_close = df.loc[df["Close"].idxmin()]

print("===== Lowest Closing Price =====")
print(lowest_close)
```

```python
print("Date:", lowest_close["Date"])
print("Company:", lowest_close["Company"])
print("Lowest Closing Price:", lowest_close["Close"])
```

```python
plt.figure(figsize=(8,5))

plt.bar(
    lowest_close["Company"],
    lowest_close["Close"]
)

plt.title("Lowest Closing Price")
plt.xlabel("Company")
plt.ylabel("Closing Price")

plt.grid(axis="y")

plt.show()
```

```python
df["Month"] = df["Date"].dt.month_name()

print(df[["Date", "Month"]].head())
```

```python
monthly_close = df.groupby("Month")["Close"].mean()

print(monthly_close)
```

```python
plt.figure(figsize=(12,6))

plt.bar(monthly_close.index, monthly_close.values)

plt.title("Monthly Average Closing Price")
plt.xlabel("Month")
plt.ylabel("Average Closing Price")

plt.xticks(rotation=45)

plt.grid(axis="y")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.histplot(
    data=df,
    x="Close",
    bins=30,
    kde=True
)

plt.title("Distribution of Closing Prices")
plt.xlabel("Closing Price")
plt.ylabel("Frequency")

plt.grid(True)

plt.show()
```

```python
selected_columns = [
    "Open",
    "High",
    "Low",
    "Close",
    "Volume",
    "Daily_Return"
]

sns.pairplot(df[selected_columns])

plt.show()
```
