```python


import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


class SalesDataAnalyzer:
    """
    Encapsulates all functionality required to load, clean,
    analyze and visualize sales data.
    """

    
    def __init__(self, file_path: str = None):
        """Constructor - optionally loads data if a file_path is given."""
        self.data: pd.DataFrame = pd.DataFrame()
        self.last_figure = None          
        if file_path:
            self.load_data(file_path)
        print("SalesDataAnalyzer object created.")

    def __del__(self):
        """Destructor - performs cleanup when the object is destroyed."""
        try:
            plt.close('all')
        except Exception:
            pass
        print("SalesDataAnalyzer object destroyed. Cleanup complete.")


    def load_data(self, file_path: str) -> None:
        """Load data from a CSV file into self.data."""
        try:
            self.data = pd.read_csv(file_path)
            print(f"Dataset loaded successfully from '{file_path}'!")
        except FileNotFoundError:
            print(f"Error: File '{file_path}' not found.")
        except pd.errors.EmptyDataError:
            print("Error: The file is empty.")
        except Exception as e:
            print(f"An unexpected error occurred while loading data: {e}")

    
    def explore_data(self) -> None:
        """Sub-menu to display basic information about the dataset."""
        if self.data.empty:
            print("No data loaded. Please load a dataset first.")
            return

        while True:
            print("\n== Explore Data ==")
            print("1. Display the first 5 rows")
            print("2. Display the last 5 rows")
            print("3. Display column names")
            print("4. Display data types")
            print("5. Display basic info")
            print("6. Back to main menu")
            choice = input("Enter your choice: ").strip()

            match choice:
                case "1":
                    print(self.data.head())
                case "2":
                    print(self.data.tail())
                case "3":
                    print(list(self.data.columns))
                case "4":
                    print(self.data.dtypes)
                case "5":
                    print(self.data.info())
                    print("\nDescribe:\n", self.data.describe(include='all'))
                case "6":
                    break
                case _:
                    print("Invalid choice. Please try again.")

    
    def clean_data(self) -> None:
        """Sub-menu to handle missing values and perform data cleaning."""
        if self.data.empty:
            print("No data loaded. Please load a dataset first.")
            return

        while True:
            print("\n== Handle Missing Data ==")
            print("1. Display rows with missing values")
            print("2. Fill missing values with mean")
            print("3. Drop rows with missing values")
            print("4. Replace missing values with a specific value")
            print("5. Back to main menu")
            choice = input("Enter your choice: ").strip()

            match choice:
                case "1":
                    missing = self.data[self.data.isnull().any(axis=1)]
                    if missing.empty:
                        print("No missing values found in the dataset!")
                    else:
                        print(missing)
                case "2":
                    numeric_cols = self.data.select_dtypes(include=np.number).columns
                    self.data[numeric_cols] = self.data[numeric_cols].apply(
                        lambda col: col.fillna(col.mean())
                    )
                    print("Missing numeric values filled with column mean.")
                case "3":
                    before = len(self.data)
                    self.data = self.data.dropna()
                    print(f"Dropped {before - len(self.data)} rows containing missing values.")
                case "4":
                    value = input("Enter the value to replace missing data with: ")
                    self.data = self.data.fillna(value)
                    print(f"Missing values replaced with '{value}'.")
                case "5":
                    break
                case _:
                    print("Invalid choice. Please try again.")

    
    def numpy_operations(self) -> None:
        """Demonstrate creation, indexing and slicing of Numpy arrays."""
        if self.data.empty:
            print("No data loaded. Please load a dataset first.")
            return

        numeric_cols = self.data.select_dtypes(include=np.number).columns.tolist()
        if not numeric_cols:
            print("No numeric columns available for Numpy operations.")
            return

        print(f"Numeric columns available: {numeric_cols}")
        col = input(f"Enter column name to convert to a Numpy array {numeric_cols}: ").strip()

        if col not in numeric_cols:
            print("Invalid column name.")
            return

        arr = self.data[col].to_numpy()
        print(f"\nNumpy array for '{col}':\n{arr}")
        print(f"Shape: {arr.shape}, Dtype: {arr.dtype}")

        print(f"\nFirst element (indexing arr[0]): {arr[0]}")
        print(f"Last element (indexing arr[-1]): {arr[-1]}")
        print(f"First 3 elements (slicing arr[:3]): {arr[:3]}")
        print(f"Every other element (slicing arr[::2]): {arr[::2]}")

    
    def mathematical_operations(self) -> None:
        """Perform element-wise mathematical operations on numeric data."""
        if self.data.empty:
            print("No data loaded. Please load a dataset first.")
            return

        numeric_cols = self.data.select_dtypes(include=np.number).columns.tolist()
        if not numeric_cols:
            print("No numeric columns available.")
            return

        print(f"Numeric columns: {numeric_cols}")
        col = input("Enter column name to perform operations on: ").strip()
        if col not in numeric_cols:
            print("Invalid column name.")
            return

        arr = self.data[col].to_numpy()

        print("\n1. Add a constant")
        print("2. Subtract a constant")
        print("3. Multiply by a constant")
        print("4. Divide by a constant")
        choice = input("Enter your choice: ").strip()

        try:
            constant = float(input("Enter the constant value: "))
        except ValueError:
            print("Invalid numeric value.")
            return

        match choice:
            case "1":
                result = arr + constant
            case "2":
                result = arr - constant
            case "3":
                result = arr * constant
            case "4":
                if constant == 0:
                    print("Cannot divide by zero.")
                    return
                result = arr / constant
            case _:
                print("Invalid choice.")
                return

        print(f"\nResult:\n{result}")

    def combine_data(self, other_dataframe: pd.DataFrame, how: str = "concat",
                      on: str = None) -> None:
        """Combine the current DataFrame with another DataFrame."""
        if self.data.empty:
            print("No data loaded. Please load a dataset first.")
            return
        try:
            match how:
                case "concat":
                    self.data = pd.concat([self.data, other_dataframe], ignore_index=True)
                case "merge":
                    self.data = pd.merge(self.data, other_dataframe, on=on, how="inner")
                case "join":
                    self.data = self.data.join(other_dataframe.set_index(on), on=on)
                case _:
                    print("Unknown combine method. Use 'concat', 'merge' or 'join'.")
                    return
            print(f"DataFrames combined successfully using '{how}'.")
        except Exception as e:
            print(f"Error while combining DataFrames: {e}")

    def combine_data_menu(self) -> None:
        """Sub-menu wrapper to combine current data with another CSV file."""
        if self.data.empty:
            print("No data loaded. Please load a dataset first.")
            return
        path = input("Enter path of CSV file to combine with current data: ").strip()
        if not os.path.exists(path):
            print("File not found.")
            return
        other_df = pd.read_csv(path)

        print("1. Concat  2. Merge  3. Join")
        choice = input("Enter your choice: ").strip()
        method_map = {"1": "concat", "2": "merge", "3": "join"}
        method = method_map.get(choice)
        if not method:
            print("Invalid choice.")
            return

        on_col = None
        if method in ("merge", "join"):
            on_col = input("Enter the common column name to combine on: ").strip()

        self.combine_data(other_df, how=method, on=on_col)

    def split_data(self) -> dict:
        """Split the DataFrame into multiple DataFrames based on a chosen column."""
        if self.data.empty:
            print("No data loaded. Please load a dataset first.")
            return {}

        print(f"Available columns: {list(self.data.columns)}")
        col = input("Enter column name to split data by: ").strip()
        if col not in self.data.columns:
            print("Invalid column name.")
            return {}

        groups = {key: grp for key, grp in self.data.groupby(col)}
        for key, grp in groups.items():
            print(f"\n--- Group: {key} ({len(grp)} rows) ---")
            print(grp.head())
        return groups

    
    def search_sort_filter(self) -> None:
        """Sub-menu implementing search, sort and filter functionalities."""
        if self.data.empty:
            print("No data loaded. Please load a dataset first.")
            return

        while True:
            print("\n== Search / Sort / Filter ==")
            print("1. Search for a specific value in a column")
            print("2. Sort data by a column")
            print("3. Filter data based on a condition")
            print("4. Back to main menu")
            choice = input("Enter your choice: ").strip()

            match choice:
                case "1":
                    col = input(f"Enter column name {list(self.data.columns)}: ").strip()
                    if col not in self.data.columns:
                        print("Invalid column name.")
                        continue
                    value = input("Enter value to search for: ").strip()
                    result = self.data[self.data[col].astype(str) == value]
                    print(result if not result.empty else "No matching records found.")

                case "2":
                    col = input(f"Enter column name to sort by {list(self.data.columns)}: ").strip()
                    if col not in self.data.columns:
                        print("Invalid column name.")
                        continue
                    order = input("Sort ascending? (y/n): ").strip().lower()
                    ascending = order != "n"
                    print(self.data.sort_values(by=col, ascending=ascending))

                case "3":
                    col = input(f"Enter column name to filter on {list(self.data.columns)}: ").strip()
                    if col not in self.data.columns:
                        print("Invalid column name.")
                        continue
                    op = input("Enter operator (==, !=, >, <, >=, <=): ").strip()
                    value = input("Enter value: ").strip()

                    try:
                        if np.issubdtype(self.data[col].dtype, np.number):
                            value = float(value)
                    except ValueError:
                        pass

                    try:
                        match op:
                            case "==":
                                result = self.data[self.data[col] == value]
                            case "!=":
                                result = self.data[self.data[col] != value]
                            case ">":
                                result = self.data[self.data[col] > value]
                            case "<":
                                result = self.data[self.data[col] < value]
                            case ">=":
                                result = self.data[self.data[col] >= value]
                            case "<=":
                                result = self.data[self.data[col] <= value]
                            case _:
                                print("Invalid operator.")
                                continue
                        print(result if not result.empty else "No matching records found.")
                    except Exception as e:
                        print(f"Error applying filter: {e}")

                case "4":
                    break
                case _:
                    print("Invalid choice. Please try again.")

    def aggregate_functions(self) -> None:
        """Apply aggregating functions like sum, mean, count etc."""
        if self.data.empty:
            print("No data loaded. Please load a dataset first.")
            return

        numeric_cols = self.data.select_dtypes(include=np.number).columns.tolist()
        if not numeric_cols:
            print("No numeric columns available.")
            return

        print(f"Numeric columns: {numeric_cols}")
        col = input("Enter column name to aggregate: ").strip()
        if col not in numeric_cols:
            print("Invalid column name.")
            return

        print("1. Sum  2. Mean  3. Count  4. Min  5. Max  6. All of the above")
        choice = input("Enter your choice: ").strip()

        series = self.data[col]
        match choice:
            case "1":
                print(f"Sum: {series.sum()}")
            case "2":
                print(f"Mean: {series.mean()}")
            case "3":
                print(f"Count: {series.count()}")
            case "4":
                print(f"Min: {series.min()}")
            case "5":
                print(f"Max: {series.max()}")
            case "6":
                print(series.agg(['sum', 'mean', 'count', 'min', 'max']))
            case _:
                print("Invalid choice.")


    def statistical_analysis(self) -> None:
        """Perform statistical computations: std, var, percentiles, describe."""
        if self.data.empty:
            print("No data loaded. Please load a dataset first.")
            return

        numeric_cols = self.data.select_dtypes(include=np.number).columns.tolist()
        if not numeric_cols:
            print("No numeric columns available.")
            return

        print(f"Numeric columns: {numeric_cols}")
        col = input("Enter column name for statistical analysis: ").strip()
        if col not in numeric_cols:
            print("Invalid column name.")
            return

        series = self.data[col]
        print(f"\n-- Statistics for '{col}' --")
        print(f"Describe:\n{series.describe()}")
        print(f"Standard Deviation (std): {series.std()}")
        print(f"Variance (var): {series.var()}")
        print(f"25th percentile: {series.quantile(0.25)}")
        print(f"50th percentile (median): {series.quantile(0.50)}")
        print(f"75th percentile: {series.quantile(0.75)}")

    
    def create_pivot_table(self) -> None:
        """Generate a pivot table for data summarization."""
        if self.data.empty:
            print("No data loaded. Please load a dataset first.")
            return

        print(f"Available columns: {list(self.data.columns)}")
        index_col = input("Enter column name for pivot table index: ").strip()
        columns_col = input("Enter column name for pivot table columns (optional, press Enter to skip): ").strip()
        values_col = input("Enter column name for pivot table values: ").strip()
        agg_func = input("Enter aggregation function (sum/mean/count) [default sum]: ").strip() or "sum"

        try:
            pivot = pd.pivot_table(
                self.data,
                index=index_col if index_col else None,
                columns=columns_col if columns_col else None,
                values=values_col if values_col else None,
                aggfunc=agg_func,
            )
            print("\nPivot Table:\n", pivot)
        except Exception as e:
            print(f"Error creating pivot table: {e}")

    
    def visualize_data(self) -> None:
        """Sub-menu to create various plots using Matplotlib and Seaborn."""
        if self.data.empty:
            print("No data loaded. Please load a dataset first.")
            return

        while True:
            print("\n== Data Visualization ==")
            print("1. Bar Plot")
            print("2. Line Plot")
            print("3. Scatter Plot")
            print("4. Pie Chart")
            print("5. Histogram")
            print("6. Stack Plot")
            print("7. Seaborn Heatmap")
            print("8. Seaborn Box Plot")
            print("9. Multiple Subplots")
            print("10. Back to main menu")
            choice = input("Enter your choice: ").strip()

            match choice:
                case "1":
                    x = input("Enter x-axis column name: ").strip()
                    y = input("Enter y-axis column name: ").strip()
                    if x not in self.data.columns or y not in self.data.columns:
                        print("Invalid column name(s).")
                        continue
                    print("Generating bar plot...")
                    fig, ax = plt.subplots(figsize=(8, 5))
                    ax.bar(self.data[x].astype(str), self.data[y])
                    ax.set_xlabel(x)
                    ax.set_ylabel(y)
                    ax.set_title(f"{y} by {x}")
                    ax.legend([y])
                    plt.xticks(rotation=45)
                    self.last_figure = fig
                    plt.tight_layout()
                    plt.show()
                    print("Bar plot displayed successfully!")

                case "2":
                    x = input("Enter x-axis column name: ").strip()
                    y = input("Enter y-axis column name: ").strip()
                    if x not in self.data.columns or y not in self.data.columns:
                        print("Invalid column name(s).")
                        continue
                    print("Generating line plot...")
                    fig, ax = plt.subplots(figsize=(8, 5))
                    ax.plot(self.data[x], self.data[y], marker='o')
                    ax.set_xlabel(x)
                    ax.set_ylabel(y)
                    ax.set_title(f"{y} over {x}")
                    ax.legend([y])
                    self.last_figure = fig
                    plt.tight_layout()
                    plt.show()
                    print("Line plot displayed successfully!")

                case "3":
                    x = input("Enter x-axis column name: ").strip()
                    y = input("Enter y-axis column name: ").strip()
                    if x not in self.data.columns or y not in self.data.columns:
                        print("Invalid column name(s).")
                        continue
                    print("Generating scatter plot...")
                    fig, ax = plt.subplots(figsize=(8, 5))
                    ax.scatter(self.data[x], self.data[y])
                    ax.set_xlabel(x)
                    ax.set_ylabel(y)
                    ax.set_title(f"{y} vs {x}")
                    self.last_figure = fig
                    plt.tight_layout()
                    plt.show()
                    print("Scatter plot displayed successfully!")

                case "4":
                    col = input("Enter column name for pie chart (categorical): ").strip()
                    val = input("Enter column name for values: ").strip()
                    if col not in self.data.columns or val not in self.data.columns:
                        print("Invalid column name(s).")
                        continue
                    print("Generating pie chart...")
                    grouped = self.data.groupby(col)[val].sum()
                    fig, ax = plt.subplots(figsize=(7, 7))
                    ax.pie(grouped, labels=grouped.index, autopct='%1.1f%%')
                    ax.set_title(f"{val} distribution by {col}")
                    self.last_figure = fig
                    plt.tight_layout()
                    plt.show()
                    print("Pie chart displayed successfully!")

                case "5":
                    col = input("Enter column name for histogram: ").strip()
                    if col not in self.data.columns:
                        print("Invalid column name.")
                        continue
                    print("Generating histogram...")
                    fig, ax = plt.subplots(figsize=(8, 5))
                    ax.hist(self.data[col], bins=10, edgecolor='black')
                    ax.set_xlabel(col)
                    ax.set_ylabel("Frequency")
                    ax.set_title(f"Distribution of {col}")
                    self.last_figure = fig
                    plt.tight_layout()
                    plt.show()
                    print("Histogram displayed successfully!")

                case "6":
                    x = input("Enter x-axis column name: ").strip()
                    y_cols = input("Enter y-axis numeric column names (comma separated): ").strip().split(',')
                    y_cols = [c.strip() for c in y_cols]
                    if x not in self.data.columns or not all(c in self.data.columns for c in y_cols):
                        print("Invalid column name(s).")
                        continue
                    print("Generating stack plot...")
                    fig, ax = plt.subplots(figsize=(8, 5))
                    ax.stackplot(self.data[x], *[self.data[c] for c in y_cols], labels=y_cols)
                    ax.set_xlabel(x)
                    ax.set_title("Stack Plot")
                    ax.legend(loc='upper left')
                    self.last_figure = fig
                    plt.tight_layout()
                    plt.show()
                    print("Stack plot displayed successfully!")

                case "7":
                    numeric_df = self.data.select_dtypes(include=np.number)
                    if numeric_df.empty:
                        print("No numeric columns available for heatmap.")
                        continue
                    print("Generating seaborn heatmap...")
                    fig, ax = plt.subplots(figsize=(8, 6))
                    sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', ax=ax)
                    ax.set_title("Correlation Heatmap")
                    self.last_figure = fig
                    plt.tight_layout()
                    plt.show()
                    print("Heatmap displayed successfully!")

                case "8":
                    x = input("Enter categorical column name (x-axis): ").strip()
                    y = input("Enter numeric column name (y-axis): ").strip()
                    if x not in self.data.columns or y not in self.data.columns:
                        print("Invalid column name(s).")
                        continue
                    print("Generating seaborn box plot...")
                    fig, ax = plt.subplots(figsize=(8, 5))
                    sns.boxplot(x=x, y=y, data=self.data, ax=ax)
                    ax.set_title(f"Box Plot of {y} by {x}")
                    self.last_figure = fig
                    plt.tight_layout()
                    plt.show()
                    print("Box plot displayed successfully!")

                case "9":
                    numeric_cols = self.data.select_dtypes(include=np.number).columns.tolist()
                    if len(numeric_cols) < 2:
                        print("Not enough numeric columns for subplots.")
                        continue
                    print("Generating multiple subplots...")
                    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
                    axes[0].hist(self.data[numeric_cols[0]], bins=10, edgecolor='black')
                    axes[0].set_title(f"Histogram of {numeric_cols[0]}")
                    axes[1].boxplot(self.data[numeric_cols[1]])
                    axes[1].set_title(f"Box plot of {numeric_cols[1]}")
                    self.last_figure = fig
                    plt.tight_layout()
                    plt.show()
                    print("Subplots displayed successfully!")

                case "10":
                    break
                case _:
                    print("Invalid choice. Please try again.")

    def save_visualization(self) -> None:
        """Save the last generated plot/figure to a file."""
        if self.last_figure is None:
            print("No visualization available to save. Please generate a plot first.")
            return
        filename = input("Enter file name to save the plot (e.g., plot.png): ").strip()
        try:
            self.last_figure.savefig(filename)
            print(f"Visualization saved as {filename} successfully!")
        except Exception as e:
            print(f"Error saving visualization: {e}")


    @staticmethod
    def generate_synthetic_dataset(path: str = "sales_data.csv", n: int = 100) -> str:
        """Generate a synthetic sales dataset and save it as CSV."""
        rng = np.random.default_rng(42)
        products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
        regions = ['North', 'South', 'East', 'West Coast', 'Central']

        dates = pd.date_range(start="2023-01-01", periods=n, freq="D")
        df = pd.DataFrame({
            "Date": dates,
            "Product": rng.choice(products, n),
            "Region": rng.choice(regions, n),
            "Sales": rng.integers(100, 1000, n),
            "Profit": rng.integers(10, 300, n),
        })
        df.to_csv(path, index=False)
        return path



def main():
    analyzer = SalesDataAnalyzer()

    MENU = """
========== Data Analysis & Visualization Program ==========
Please select an option:
1. Load Dataset
2. Explore Data
3. Perform DataFrame Operations
4. Handle Missing Data
5. Generate Descriptive Statistics
6. Data Visualization
7. Save Visualization
8. Exit
=============================================================
"""

    while True:
        print(MENU)
        choice = input("Enter your choice: ").strip()

        match choice:
            case "1":
                print("\n== Load Dataset ==")
                print("1. Load from an existing CSV file")
                print("2. Generate a synthetic dataset")
                sub = input("Enter your choice: ").strip()
                match sub:
                    case "1":
                        path = input("Enter the path of the dataset (CSV file): ").strip()
                        analyzer.load_data(path)
                    case "2":
                        path = analyzer.generate_synthetic_dataset()
                        analyzer.load_data(path)
                    case _:
                        print("Invalid choice.")

            case "2":
                analyzer.explore_data()

            case "3":
                print("\n== Perform DataFrame Operations ==")
                print("1. Numpy Array Creation / Indexing / Slicing")
                print("2. Mathematical Operations")
                print("3. Combine Data (with another CSV)")
                print("4. Split Data")
                print("5. Search / Sort / Filter")
                print("6. Aggregate Functions")
                print("7. Create Pivot Table")
                print("8. Back to main menu")
                sub = input("Enter your choice: ").strip()
                match sub:
                    case "1":
                        analyzer.numpy_operations()
                    case "2":
                        analyzer.mathematical_operations()
                    case "3":
                        analyzer.combine_data_menu()
                    case "4":
                        analyzer.split_data()
                    case "5":
                        analyzer.search_sort_filter()
                    case "6":
                        analyzer.aggregate_functions()
                    case "7":
                        analyzer.create_pivot_table()
                    case "8":
                        pass
                    case _:
                        print("Invalid choice.")

            case "4":
                analyzer.clean_data()

            case "5":
                analyzer.statistical_analysis()

            case "6":
                analyzer.visualize_data()

            case "7":
                analyzer.save_visualization()

            case "8":
                print("Exiting program. Thank you for using Sales Data Analyzer!")
                del analyzer
                break

            case _:
                print("Invalid choice. Please enter a number between 1 and 8.")


if __name__ == "__main__":
    main()
```
