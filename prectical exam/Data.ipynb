```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


class RetailAnalyzer:

    def __init__(self):
        self.data = None

    # Step 1 : Load Dataset
    def load_data(self):

        file_path = "retail_sales.csv"

        if os.path.exists(file_path):

            self.data = pd.read_csv(file_path)

            print("Dataset Loaded Successfully")

        else:

            print("File Not Found")

    # Step 2 : Display Dataset
    def display_data(self):

        if self.data is not None:

            print("\nFirst 5 Records\n")

            print(self.data.head())

        else:

            print("No data loaded.")

    # Step 3 : Check Missing Values
    def check_missing(self):

        if self.data is not None:

            print("\nMissing Values\n")

            print(self.data.isnull().sum())

        else:

            print("No data loaded.")

    # Step 4 : Fill Missing Values
    def fill_missing(self):

        if self.data is not None:

            self.data["Price"] = self.data["Price"].fillna(self.data["Price"].mean())

            self.data["Quantity Sold"] = self.data["Quantity Sold"].fillna(
                self.data["Quantity Sold"].mean())

            self.data["Total Sales"] = self.data["Total Sales"].fillna(
                self.data["Total Sales"].mean())

            self.data["Product"] = self.data["Product"].fillna("Unknown")

            self.data["Category"] = self.data["Category"].fillna("Unknown")

            self.data["Date"] = self.data["Date"].fillna("2025-01-01")

            print("\nMissing Values Filled Successfully")

        else:

            print("No data loaded.")

    # Step 5 : Calculate Metrics
    def calculate_metrics(self):

        if self.data is not None:

            total_sales = self.data["Total Sales"].sum()

            average_sales = self.data["Total Sales"].mean()

            maximum_sale = self.data["Total Sales"].max()

            minimum_sale = self.data["Total Sales"].min()

            popular_product = self.data["Product"].mode()[0]

            print("\n========== SALES REPORT ==========\n")

            print("Total Sales :", total_sales)

            print("Average Sales :", average_sales)

            print("Maximum Sale :", maximum_sale)

            print("Minimum Sale :", minimum_sale)

            print("Most Popular Product :", popular_product)

        else:

            print("No data loaded.")

        # Step 6 : NumPy Analysis
    def numpy_analysis(self):

        if self.data is not None:

            sales = np.array(self.data["Total Sales"])

            print("\n========== NUMPY ANALYSIS ==========\n")

            print("Total Sales :", np.sum(sales))
            print("Average Sales :", np.mean(sales))
            print("Maximum Sale :", np.max(sales))
            print("Minimum Sale :", np.min(sales))
            print("Standard Deviation :", np.std(sales))

        else:

            print("No data loaded.")


    # Step 7 : Sales Percentage
    def sales_percentage(self):

        if self.data is not None:

            sales = np.array(self.data["Total Sales"])

            percentage = (sales / np.sum(sales)) * 100

            self.data["Sales Percentage"] = percentage

            print("\nSales Percentage Added Successfully\n")

            print(self.data[["Product", "Total Sales", "Sales Percentage"]].head())

        else:

            print("No data loaded.")


    # Step 8 : Filter Data
    def filter_data(self):

        if self.data is not None:

            category = input("Enter Category : ")

            filtered_data = self.data[self.data["Category"] == category]

            if not filtered_data.empty:

                print("\nFiltered Data\n")

                print(filtered_data)

            else:

                print("No Record Found.")

        else:

            print("No data loaded.")


    # Step 9 : Dataset Summary
    def summary(self):

        if self.data is not None:

            print("\n========== DATASET SUMMARY ==========\n")

            print("Shape :", self.data.shape)

            print("\nColumns :")
            print(self.data.columns)

            print("\nData Types :")
            print(self.data.dtypes)

            print("\nInformation :")
            self.data.info()

            print("\nStatistical Summary :")
            print(self.data.describe())

        else:

            print("No data loaded.")


    # Step 10 : Display Updated Dataset
    def display_updated_data(self):

        if self.data is not None:

            print("\nUpdated Dataset\n")

            print(self.data.head(10))

        else:

            print("No data loaded.")

        # Step 11 : Data Visualization
    def visualize_data(self):

        if self.data is not None:

            # Bar Chart
            plt.figure(figsize=(8,5))

            category_sales = self.data.groupby("Category")["Total Sales"].sum()

            category_sales.plot(kind="bar", color="skyblue")

            plt.title("Total Sales by Category")

            plt.xlabel("Category")

            plt.ylabel("Total Sales")

            plt.show()


            # Line Chart
            plt.figure(figsize=(8,5))

            plt.plot(self.data["Total Sales"], marker="o")

            plt.title("Total Sales Trend")

            plt.xlabel("Index")

            plt.ylabel("Total Sales")

            plt.grid(True)

            plt.show()


            # Heatmap
            plt.figure(figsize=(8,6))

            sns.heatmap(self.data.corr(numeric_only=True), annot=True, cmap="Blues")

            plt.title("Correlation Heatmap")

            plt.show()

        else:

            print("No data loaded.")

    # Step 12 : Menu Driven Program

while True:

    print("\n========== RETAIL SALES ANALYZER ==========")
    print("1. Load Dataset")
    print("2. Display Dataset")
    print("3. Check Missing Values")
    print("4. Fill Missing Values")
    print("5. Calculate Sales Metrics")
    print("6. NumPy Analysis")
    print("7. Sales Percentage")
    print("8. Filter Data")
    print("9. Dataset Summary")
    print("10. Display Updated Dataset")
    print("11. Data Visualization")
    print("12. Exit")

    choice = int(input("\nEnter Your Choice : "))

    if choice == 1:

        analyzer.load_data()

    elif choice == 2:

        analyzer.display_data()

    elif choice == 3:

        analyzer.check_missing()

    elif choice == 4:

        analyzer.fill_missing()

    elif choice == 5:

        analyzer.calculate_metrics()

    elif choice == 6:

        analyzer.numpy_analysis()

    elif choice == 7:

        analyzer.sales_percentage()

    elif choice == 8:

        analyzer.filter_data()

    elif choice == 9:

        analyzer.summary()

    elif choice == 10:

        analyzer.display_updated_data()

    elif choice == 11:

        analyzer.visualize_data()

    elif choice == 12:

        print("\nThank You for Using Retail Sales Analyzer!")
        break

    else:

        print("\nInvalid Choice! Please Try Again.")

# ----------------------------
# Object Creation
# ----------------------------

analyzer = RetailAnalyzer()

# Step 1
analyzer.load_data()

# Step 2
analyzer.display_data()

# Step 3
analyzer.check_missing()

# Step 4
analyzer.fill_missing()

print("\nAfter Filling Missing Values\n")

analyzer.check_missing()

# Step 5
analyzer.calculate_metrics()

# Step 6
analyzer.numpy_analysis()

# Step 7
analyzer.sales_percentage()

# Step 8
analyzer.filter_data()

# Step 9
analyzer.summary()

# Step 10
analyzer.display_updated_data()

# Step 11
analyzer.visualize_data()
```
